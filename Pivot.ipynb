```python

import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
pivot1 = titanic.pivot_table(index='sex', columns='class', aggfunc={'survived':sum, 'fare':'mean', 'alone' : sum})
pivot1
```

```python
print(pivot1.columns)
pivot1.iloc[:, pivot1.columns.get_level_values(0)=='survived'].sum().sum()
```

```python
pivot2 = titanic.pivot_table(index='sex', columns=['class','embark_town'], aggfunc={'survived':sum, 'fare':'mean', 'alone' : sum})
print (pivot2.columns.nlevels)
pivot2
```

```python

pivot3 = pivot2.stack(list(range(pivot2.columns.nlevels))).unstack(level=1).reset_index()
print(pivot3)
pivot3[['survived']].sum()
```

```python
pivot5 = titanic.pivot_table('survived', index='sex', columns='class')
pivot5
```

```python
age = pd.cut(titanic['age'], [0, 18, 80])
pivot6 = titanic.pivot_table('survived', ['sex', age,'embark_town'], 'class')
pivot6
```

```python
fare = pd.qcut(titanic['fare'], 2)
pivot7 = titanic.pivot_table('survived', ['sex', age], [fare, 'class'])
pivot7
```

```python
pivot8 = titanic.pivot_table(index='sex', columns='class',
                    aggfunc={'survived':sum, 'fare':'mean'})
pivot8
```

```python
pivot9 = titanic.pivot_table(index='sex', columns='class', margins=True)
pivot9
```

```python
df = pd.DataFrame({'player': ['A', 'B', 'C', 'D'], 'year1': [12, 15, 19, 19], 'year2' : [22, 29, 18, 12]})
df
```

```python
df_stack = df.set_index('player').stack().reset_index()
df_stack.columns = ['player', 'year', 'value']
df_stack
```

```python
df_melt = pd.melt(df, id_vars=['player'], value_vars=['year1', 'year2'])
df_melt.columns = ['player', 'year', 'value']
df_melt
```

```python
df_pivot = df_melt.pivot(index='player', columns='year', values='value')
df_pivot.reset_index(inplace=True)
df_pivot
```

```python
# Calculate a Pandas Weighted Average Using Numpy
import pandas as pd
import numpy as np

df = pd.DataFrame.from_dict({
    'NumCourses': [3, 2, 4, 6, 2],
    'Grades': [90, 85, 95, 85, 70]
})

weighted_average = np.average(a=df['Grades'], weights=df['NumCourses'])
print(weighted_average)


```
